```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o dataset tratado
df = pd.read_excel('testetratado.xlsx')

#
```

```python
# Análise 1: Marcas mais populares
marca_counts = df['marca'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=marca_counts.index, y=marca_counts.values, palette='viridis')
plt.title('Quantidade de Produtos por Marca')
plt.xlabel('Marca')
plt.ylabel('Quantidade')

plt.tight_layout()
plt.show()
```

```python
# Análise 2: Cores mais comuns
cor_counts = df['cor'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(8, 6))
sns.barplot(x=cor_counts.index, y=cor_counts.values, palette='pastel')
plt.title('Cores dos Produtos')
plt.xlabel('Cor')
plt.ylabel('Quantidade')

plt.tight_layout()
plt.show()
```

```python
# Análise 3: % de produtos com lente
lente_counts = df['tem_lente'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(lente_counts, labels=['Tem Lente', 'Não Tem Lente'], autopct='%1.1f%%', colors=['#66b3ff','#ff9999'])
plt.title('Produtos com Lente')
plt.tight_layout()
plt.show()
```

```python
# Análise 4: Preço médio por marca (Se tiver preço numerico)
def limpar_preco(preco):
    preco = preco.replace('R$', '').replace('.', '').replace(',', '.').strip()
    try:
        return float(preco)
    except:
        return None

df['preco_num'] = df['preco'].apply(limpar_preco)

preco_medio_marca = df.groupby('marca')['preco_num'].mean().dropna().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=preco_medio_marca.index, y=preco_medio_marca.values, palette='rocket')
plt.title('Preço Médio por Marca')
plt.xlabel('Marca')
plt.ylabel('Preço Médio (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print('-' * 40)
print(f'Total de produtos analisados: {len(df)}')
print(f'Marcas encontradas: {df["marca"].nunique()}')
print(f'Cores distintas: {df["cor"].nunique()}')
print(f'Percentual de produtos com lente: {lente_counts[True]/lente_counts.sum()*100:.2f}%')
print('-' * 40)
```
